```python
#zad1

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, Eq, solve

_x = np.array([0.4, 0.8, 1.2, 1.6, 2, 2.3])
_y = np.array([800, 975, 1500, 1950, 2900, 3600])

a, b, x = symbols('a, b, x')

def f(a, b, x):
    return a + b*x

def F(a, b):
    sum = 0
    for i in range (0, _x.size):
        sum += (f(a, b, _x[i]) - np.log10(_y[i]))**2
    return sum

eq = [Eq(diff(F(a, b), a), 0), Eq(diff(F(a, b), b), 0)]
sol = solve(eq)

def func(x):
    return 10 ** (sol[a] + sol[b]*x)

x_axis = np.linspace(0, 2.5, 100)
plt.plot(x_axis, func(x_axis), color='violet')
plt.scatter(_x, _y)
plt.show()
```

```python
#zad2 

import numpy as np
import matplotlib.pyplot as plt

_x = np.array([0.2, 0.5, 0.8, 1.2, 1.7, 2, 2.3])
_y = np.array([500, 700, 1000, 1200, 2200, 2650, 3750])
A = np.zeros([3, 3])
B = np.zeros([3])

for i in range(3):
    for j in range(3):
       for k in range(_x.size):
           A[i][j] +=  _x[k]**(i+j)

for i in range(3):
    for k in range(_x.size):
           B[i] +=  _y[k] * (_x[k]**i)

sol = np.linalg.solve(A, B)

def f(x):
    return sol[2]*(x**2) + sol[1]*x + sol[0]

x_axis = np.linspace(0, 2.5, 100)
plt.scatter(_x, _y)
plt.plot(x_axis, f(x_axis), color='violet')
plt.show()
```

```python
#zad3

import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, Eq, solve

_x = np.array([2.5, 3.5, 5, 6, 7.5, 10, 12.5, 15, 17.5, 20])
_y = np.array([13, 11, 8.5, 8.2, 7, 6.2, 5.2, 4.8, 4.6, 4.3])

a, b, x = symbols('a, b, x')

def f(a, b, x):
    return a + b*np.log(x)

def F(a, b):
    sum = 0
    for i in range (0, _x.size):
        sum += (f(a, b, _x[i]) - np.log(_y[i]))**2
    return sum

eq = [Eq(diff(F(a, b), a), 0), Eq(diff(F(a, b), b), 0)]
sol = solve(eq)

def func(x):
    return np.exp(float(sol[a]) + (np.log(x)*float(sol[b])))

plt.scatter(_x, _y)
x_axis = np.linspace(2, 20, 100)
plt.plot(x_axis, func(x_axis), color='violet')
plt.scatter(9, func(9), color='r')
plt.show()
```

```python
#zad4 
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, Eq, solve

_x = np.array([0.1, 0.2, 0.4, 0.6, 0.9, 1.3, 1.5, 1.7, 1.8])
_y = np.array([0.75, 1.25, 1.45, 1.25, 0.85, 0.55, 0.35, 0.28, 0.18])
a, b, x = symbols('a, b, x')

def f(a, b, x):
    return a + np.log(x) + b*x

def F(a, b):
    sum = 0
    for i in range(_x.size):
        sum += (f(a, b, _x[i]) - np.log(_y[i]))**2
    return sum

eq = [Eq(diff(F(a,b), a), 0), Eq(diff(F(a, b), b), 0)]
sol = solve(eq)

def func(x):
    return np.exp(float(sol[a]) + np.log(x) + float(sol[b])*x)

plt.scatter(_x, _y)
x_axis = np.linspace(0.0001, 2, 100)
plt.plot(x_axis, func(x_axis), color='violet')
plt.show()
```

```python
#zad7
import numpy as np
import matplotlib.pyplot as plt
from sympy import integrate, Eq, solve, sin

A1, A2, A3, x1, x2, x3, x = symbols('A1 A2 A3 x1 x2 x3 x')

eq = [Eq((A1 + A2 + A3), integrate(1, (x, -1, 1))),
      Eq((A1*x1 + A2*x2 + A3*x3), integrate(x, (x, -1, 1))),
      Eq((A1*(x1**2) + A2*(x2**2) + A3*(x3**2)), integrate((x**2), (x, -1, 1))),
      Eq((A1*(x1**3) + A2*(x2**3) + A3*(x3**3)), integrate((x**3), (x, -1, 1))),
      Eq((A1*(x1**4) + A2*(x2**4) + A3*(x3**4)), integrate((x**4), (x, -1, 1))),
      Eq((A1*(x1**5) + A2*(x2**5) + A3*(x3**5)), integrate((x**5), (x, -1, 1)))]
sol = solve(eq, A1, x1, A2, x2, A3, x3, dict=True)

A1, x1, A2, x2, A3, x3 = sol[0]

def f(x):
    return np.e**(sin(x)) - 2*(x**2)

print((integrate(f(x), (x, 0, 1))).evalf())
print((A1*f(x1) + A2*f(x2) + A3*f(x3)).evalf())
```

```python
#zad8
import numpy as np

_x = np.linspace(-1, 1, 6)

def f(x):
    return -x**4 + 1

result = 0
for i in range(1, _x.size):
    result += f(_x[i-1]) + f(_x[i])

result *= (1 - (-1)) / (2 * 5)

exact_val = integrate(-x**4 + 1, (x, -1, 1))

print(result)
print(exact_val)
```
