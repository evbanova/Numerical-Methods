```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def least_squares_matrix(x, y, n):
    matrix = np.zeros([n, n])
    for i in range(n):
        for j in range(n):
            matrix[i][j] += sum(x[t]**(i+j) for t in range(n))
    vector = np.zeros(n)
    for i in range(n):
        vector[i] = sum(y[t]*(x[t]**i) for t in range(n))
    solution = np.linalg.solve(matrix, vector)
    return solution

def f(_x, x, y, n):
    solution = least_squares_matrix(x, y, n)
    result = 0
    for i in range(n):
        result += (solution[i] * (_x ** i))
    return result    

xs = np.array([0, 1, 2, 3, 4])
ys = np.array([0, 1, 1, 2, 2])

x_axis = np.linspace(0, 5, 100)
plt.scatter(xs, ys)
plt.plot(x_axis, f(x_axis, xs, ys, 5))
plt.show()
```

```python
df = pd.read_csv('CO_2_data.csv')
x_col = df['year']
y_col = df['co2']
x_axis = np.linspace(1940, 2020, 1000)
plt.plot(x_axis, f(x_axis, x_col, y_col, x_col.size))
plt.scatter(x_col, y_col)
plt.show()
```
