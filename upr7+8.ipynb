```python
import numpy as np
import matplotlib.pyplot as plt
import math
from sympy import Symbol
_x = Symbol('x')
```

```python
#zad1 Задача 1
#В таблицата са дадени данни за развитието на бактериална популация

#           t, h	1	2	3	4	   5
#бр. клетки (x1000)	1	12	110	1037   12218
#Да се намери подходяща функция, която интерполира данните.
```

```python
nodes = np.array([1, 2, 3, 4, 5])
values = np.array([1, 12, 110, 1037, 12218])
matrix = np.zeros((5, 5))
for i in range(0,5):
    for j in range(0,5):
        matrix[i][j] = np.e**(nodes[i]*j)
sol = np.linalg.solve(matrix, values)

def p(x):
    return sol[0] + sol[1]*(np.e**x) + sol[2]*(np.e**(2*x)) + sol[3]*(np.e**(3*x)) + sol[4]*(np.e**(4*x))

x = np.linspace(5, 10, 100)
plt.plot(x, p(x))
plt.show()
```
